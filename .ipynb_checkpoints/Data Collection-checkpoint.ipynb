```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
pd.set_option('display.max_rows', None)
```

# Load Data
### Load Revenue and Pay Data

```python
p5 = ['ACC', 'SEC', 'Big Ten', 'Big 12', 'Ind', 'Pac-12']
```

```python
# Load the coaching pay data
coach_pay_link = "https://sportsdata.usatoday.com/ncaa/salaries/football/coach"
coach_pay = pd.read_html(coach_pay_link)[0]
coach_pay
```

```python
school_revenue_link = "https://sportsdata.usatoday.com/ncaa/finances"
school_revenue = pd.read_html(school_revenue_link)[0]
school_revenue
```

```python
# Merge the data together
team_money = coach_pay.merge(school_revenue, on=['School'], how='outer')
team_money
```

```python
# Clean to relevant columns and P5 programs
team_money.rename({'Conf._x': 'Conf', 'Total Pay': 'Coach Pay', 'Total Revenue': 'Revenue'}, inplace=True, axis=1)
team_money['Conf'] = team_money['Conf'].apply(lambda x: "Big Ten" if x == "Big 10" else x)
team_money['Conf'] = team_money['Conf'].apply(lambda x: "Ind" if x == "IndFBS" else x)
team_money = team_money[team_money['Conf'].isin(p5)]
team_money = team_money[~(
    ((team_money['Conf'] == 'Ind') & (team_money['School'] != 'Notre Dame')) | 
    (team_money['School'] == 'Oregon State') |
    (team_money['School'] == 'Washington State')
)]
team_money = team_money[['School', 'Conf', 'Coach Pay', 'Revenue']].sort_values('Conf').reset_index(drop=True)
team_money
```

```python
def clean_money(value):
    if isinstance(value, str):  # Ensure it's a string before processing
        value = value.replace(",", "")  # Remove commas
        value = value.lstrip("$")  # Remove leading dollar sign
        value = value.rstrip("*")  # Remove trailing asterisk
        
        if value.isdigit():  # Check if the cleaned value is purely numeric
            return int(value)
    
    return np.nan  # Return NaN for non-numeric values

# Turn pay and revenue data into integer form
team_money["Coach Pay"] = team_money["Coach Pay"].apply(clean_money)
team_money["Revenue"] = team_money["Revenue"].apply(clean_money)
```

```python
team_money
```

### Load Recruiting Data

```python
url = "https://www.on3.com/db/rankings/industry-team/football/2022/"
response = requests.get(url)

if response.status_code == 200:
    page_content = response.text
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
soup = BeautifulSoup(page_content, 'html.parser')
```

```python
# Find all the teams using their repeating div classes
teams = soup.find_all('div', class_='TeamRow_teamContainer__TTLbt')
```

```python
print(teams[0].prettify())  # Inspect one team's full HTML to debug
```

```python
data = []

# Loop through each team
for team in teams:
    
    # Extract the team name
    team_name = team.find('a', class_='MuiTypography-root MuiTypography-h5 MuiLink-root MuiLink-underlineHover TeamRow_teamName__urGek css-1rtd33f').text.strip()

    # Extract the combined "Stars" data
    stars_data = team.find('div', class_='TeamRow_starsContainer__wmDhS').text.strip()

    # Split the combined data into individual counts
    stars_split = stars_data.split('5-Stars')  # Use the delimiter to separate the data
    five_stars = int(stars_split[1].split('4-Stars')[0].strip())  # Extract number of 5-Stars
    four_stars = int(stars_split[1].split('4-Stars')[1].split('3-Stars')[0].strip())  # Extract number of 4-Stars
    three_stars = int(stars_split[1].split('3-Stars')[1].split('Total')[0].strip()  )# Extract number of 3-Stars
    total_commits = int(stars_split[1].split('Total')[1].strip())  # Extract total commits
    
    # Extract average NIL
    avg_nil = team.find('span', class_='TeamRow_averageNil___jo9s').text.strip()
    avg_nil = float(avg_nil.split('$')[1].rstrip("K")) * 1000

    # Append the data to the list
    data.append([team_name, five_stars, four_stars, three_stars, total_commits, avg_nil])

recruit_df = pd.DataFrame(data, columns=['Team Name', '5 Stars', '4 Stars', '3 Stars', 'Total Commits', 'AVG NIL'])
recruit_df
```

```python
recruit_df['Team Name'] = recruit_df['Team Name'].replace({
    'Miami': 'Miami (FL)',
    'NC State': 'North Carolina State'
})
```

### Load School Year by Year Win/Loss Data and Merge

```python
yearly_data_dict = {}

for year in range(2014, 2025):
    
    # Load the HTML table into Pandas
    win_pct_link = f"https://www.sports-reference.com/cfb/years/{year}-standings.html"
    win_pct_df = pd.read_html(win_pct_link, header=1)[0]
    
    # Clean the data to only have P5 and the specific columns to use
    win_pct_df.dropna(subset='Conf', axis=0, inplace=True)
    win_pct_df['Conf'] = win_pct_df['Conf'].apply(lambda x: x.split(' (')[0])
    win_pct_df = win_pct_df[win_pct_df['Conf'].isin(p5)]
    win_pct_df = win_pct_df[~((win_pct_df['Conf'] == 'Ind') & (win_pct_df['School'] != 'Notre Dame'))]
    win_pct_df['Pct'] = win_pct_df['Pct'].apply(lambda x: float(x) * 100)
    win_pct_df = win_pct_df[['School', 'Conf', 'W', 'L', 'Pct']].reset_index(drop=True)
    
    # Add the DF to a dictionary with the year as the key
    yearly_data_dict[year] = {'df': win_pct_df}
```

```python
last_season_team = yearly_data_dict[2024]['df'].copy()
last_season_team['School'] = last_season_team['School'].replace({
    'Pitt': 'Pittsburgh',
    'Texas Christian': 'TCU'
})
last_season_team
```

```python
merged_df = last_season_team.merge(team_money, how='outer', on=['School', 'Conf'])
merged_df
```

```python
all_data_df = merged_df.merge(recruit_df, how="outer", left_on="School", right_on="Team Name")
all_data_df.drop("Team Name", inplace=True, axis=1)
all_data_df
```

```python
all_data_df.rename(
    columns={
        '5 Stars': '5 Stars_22',
        '4 Stars': '4 Stars_22',
        '3 Stars': '3 Stars_22',
        'Total Commits': 'Total Commits_22',
        'AVG NIL': 'AVG NIL_22'    
    },
    inplace=True
)
```

```python
DATA_FILE = "all_data.xlsx"
```

```python
all_data_df.to_excel(DATA_FILE, index=False)
```

# EDA
### Reload Data
Data was updated with manual data from On3 for recruiting rankings due to Selenium packaging issue and lack of time

```python
all_data_df = pd.read_excel(DATA_FILE)
```

```python
all_data_df
```

```python
# Get win percentage difference from average per school
avg_win_pct = all_data_df['Pct'].mean()
all_data_df['Win Pct Diff'] = all_data_df['Pct'] - avg_win_pct
all_data_df
```

### Get standard deviation by year

```python
year_lst = []
stdev_lst = []
for year in yearly_data_dict:
    temp_df = yearly_data_dict[year]['df']
    stdev = temp_df['Pct'].std()
    print(f"{year}: {stdev}")
    year_lst.append(year)
    stdev_lst.append(stdev)
    yearly_data_dict[year]['stdev'] = stdev
```

```python
# Plot the Standard Deviation of win percentage over time
plt.plot(year_lst, stdev_lst, marker='o', linestyle='-')
plt.axvline(x=2020.5, color='black', linestyle='--', linewidth=2, label='NIL Implemented (Post 2020)')

# Graph organization
plt.xlabel("Year")
plt.ylabel("Standard Deviation of Win %")
plt.title("Standard Deviation of P5 Win %: NCAA Football")
plt.legend()
plt.show()
```

### Corr Heatmap

```python
import seaborn as sns

# Filter to numeric values
heatmap_df = all_data_df[['Coach Pay', 'Revenue', '5 Stars_22', '4 Stars_22', '3 Stars_22', 
                          'Total Commits_22', 'AVG NIL_22', 'Win Pct Diff']]
```

```python
# Compute correlation matrix
corr_matrix = heatmap_df.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Heatmap", fontsize=14)

# Show plot
plt.show()
```

### Conferences and Winning Percentage

```python
# Exclude "Ind" from the dataset
filtered_df = all_data_df[all_data_df['Conf'] != 'Ind']

# Group by 'Conf' and compute the average 'Win Pct Diff'
conf_avg_win_pct_diff = filtered_df.groupby('Conf')['Win Pct Diff'].mean().reset_index()

# Sort values for better visualization
conf_avg_win_pct_diff = conf_avg_win_pct_diff.sort_values(by='Win Pct Diff', ascending=False)

# Create the plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=conf_avg_win_pct_diff, x='Conf', y='Win Pct Diff', palette='coolwarm')

# Add numerical labels on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, color='black')

# Formatting
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.ylabel("Average Win Pct Diff")
plt.xlabel("Conference")
plt.title("Average Win Percentage Difference by Conference (Excluding 'Ind')")

# Show plot
plt.show()
```

### Descriptive Statistics

```python
# Compute descriptive statistics
summary_stats = heatmap_df.describe().T  # Transpose to have variables as rows

# Select relevant statistics and rename for clarity
summary_stats = summary_stats[['mean', 'std', 'min', 'max', 'count']].rename(
    columns={'mean': 'Mean', 'std': 'St. Dev', 'min': 'Min', 'max': 'Max', 'count': 'Count Non-NA'}
)

# Format numbers to avoid scientific notation and limit decimals
summary_stats = summary_stats.applymap(lambda x: f"{x:,.2f}" if isinstance(x, (int, float)) else x)

# Display the table
summary_stats
```
